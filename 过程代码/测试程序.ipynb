```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score,mean_squared_error
```

```python
model=tf.keras.models.load_model('C:/Users/14760/2025051005(10).h5')
#加载模型
data=pd.read_csv('C:/Users/14760/Desktop/cleaned_data.csv')
features=data.drop(['Ms','Hc','Hc_log'],axis=1)
target_Ms=data['Ms']
target_Hc=data['Hc_log']
target=np.array([target_Ms,target_Hc]).T
#划分目标变量
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.15,random_state=10)
#划分数据集
```

```python
y_pred = model.predict(X_test)
```

```python
r2=r2_score(y_test[:,0],y_pred[:,0]*30)
rmse=np.sqrt(mean_squared_error(y_test[:,0],y_pred[:,0]*30))
print(r2)
print(rmse)
```

```python
x=np.linspace(0,250,1000)
y=x
plt.plot(x,y)
plt.scatter(y_test[:,0],y_pred[:,0]*30)
plt.show()
```

```python
r2=r2_score(y_test[:,1],y_pred[:,1])
rmse=np.sqrt(mean_squared_error(y_test[:,1],y_pred[:,1]))
print(r2)
print(rmse)
```

```python
x=np.linspace(0,7,1000)
y=x
plt.plot(x,y)
plt.scatter(y_test[:,1],y_pred[:,1])
plt.show()
```
